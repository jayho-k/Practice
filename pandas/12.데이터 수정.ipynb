# 12.데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

## 칼럼(세로) 수정

```python
#{} 형식으로 작성한다
df['학교'].replace({'북산고':'상북고'})
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace = True)
df
```

```python
# 모두 소문자로 바꿈
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교']+ '등학교'
df
```

## 칼럼(세로) 추가

```python
df['총합'] = df['국어']+ df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail' # 결과 칼럼을 추가하고 전체 데이터는 fail 로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 총합이 400보다 큰 데이터에 대해서 결과를 pass로 업데이트
df
```

## 칼럼 삭제

```python
df.drop(columns = ['총합', '국어'. '영어', '수학'])
```

## Row 삭제

```python
df.drop(index = '4번')
```

```python
filt = df['수학'] < 80
df[filt] 
```

## Row 추가

```python
# 9번학생 추가하기
df.loc['9번'] = ['이정환','해남고등학교', 184, 90, 90, 90, 90, 90, 'Kotlin', 450, 'Pass']
df
```

## cell 수정

```python
df.loc['4번','SW특기'] = 'Python'
df
```

```python
df.loc['5번',['학교','SW특기']] = ['능남고등학교', 'C'] # 능남고등학교로 전학, c언어 할줄 앎
df
```

## 컬럼 순서 변경

```python
cols = list(df.columns)
cols
```

```python
# 결과 컬럼을 맨 앞으로 꺼내고 싶음
# 핲에꺼는 값이고 뒤에 것들은 리스트 나열하고 싶은 것들이다
# 즉 [결과] + [이름, 학교 , 키 .....] 라는 뜻이다

df = df[[cols[-1]]+ cols[0:-1]]
df
    # [cols[-1]은 값이기 때문에 리스트로 한번더 감싸준다
    #이렇게 하는 이유는 df = df['결과''등등] 할 수 있는데 다 쓰기 귀찮아서 이렇게 하는 것임
```
