# 09 bar graph

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']= False
```

```python
labels = ['강백호','서태웅', '정대만']
values = [190,187,184]

plt.barh(labels, values) # 뒤에다가 h만 붙여주면 가능하다
plt.xlim(180,195)
```

```python
# set_hatch
bar = plt.bar(labels, values, width = 0.3)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('x')
plt.ylim(180,195)
```

```python
bar = plt.bar(labels, values)
plt.ylim(180,195)
# 텍스트 써주기
# ha = 'center'는 숫자를 가운데로 맞춰주게 만드는 것
for i,rect in enumerate(bar):
    plt.text(i,rect.get_height()+0.5, values[i], ha='center')
```
