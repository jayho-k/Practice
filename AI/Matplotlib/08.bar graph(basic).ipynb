# 08.bar graph(basic)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['강백호','서태웅', '정대만']
values = [190,187,184]

plt.bar(labels, values)
```

```python
plt.bar(labels, values, color = 'g')
```

```python
labels = ['강백호','서태웅', '정대만']
values = [190,187,184]
colors = ['r','g','b']


plt.bar(labels, values, color=colors)
```

```python
labels = ['강백호','서태웅', '정대만']
values = [190,187,184]


plt.bar(labels, values, width=0.5) # width는 두께 줄이기
plt.ylim(175,195) # y축
```

```python
plt.bar(labels, values, width = 0.3)
plt.xticks(rotation=45)
plt.ylim(180,191)
# 축의 이름을 45도 회전
# y축도 가능
```

```python
labels = ['강백호','서태웅', '정대만']
values = [190,187,184]
colors = ['r','g','b']
 

plt.bar(labels, values, color=colors)
ticks = ['1번','2번','3번']
plt.xticks(labels, ticks)
```
