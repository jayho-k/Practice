# 16.multiple graph

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
df= pd.read_excel('score.xlsx')
df
```

```python
fig, axs = plt.subplots(2,2,figsize=(15,10))
fig.suptitle('m_graph')

axs[0,0].bar(df['이름'], df['국어'], label='국어점수')
axs[0,0].set_title('first')
axs[0,0].legend()
axs[0,0].set(xlabel='이름', ylabel='점수')
axs[0,0].set_facecolor('lightyellow')
axs[0,0].grid(linestyle='--', linewidth=0.5)
plt.setp(axs[0,0].get_xticklabels(), rotation=30, ha="right")


axs[0,1].plot(df['이름'], df['수학'], label='수학')
axs[0,1].plot(df['이름'], df['영어'], label='영어')
axs[0,1].legend()

axs[1,0].barh(df['이름'], df['키'], label='키')
axs[1,0].legend()

axs[1,1].plot(df['이름'], df['사회'], color='g')
plt.show()
```
