## 데이터 제외하기 dropna

```python
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
# NaN 데이터를 포함된 정보를 삭제한다
df.dropna(inplace = True)
df
```

```python
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

- axis : index or coiumns (가로 세로 중에 고르는 것)
- how  :any or all

```python
df.dropna(axis = 'index', how = 'any') # NaN 가 하나라도 있는 가로 삭제
```

```python
df.dropna(axis = 'columns', how = 'any')
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis = 'columns', how = 'all')  # 세로가 전체가 nan일때 삭제 하는 것
```
