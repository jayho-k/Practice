# 14.pie chart

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
v = [30, 25, 20, 13, 10, 2]
labels = ['python','java','c','c++','go','etc']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.05]*6

plt.pie(v,labels=labels, autopct='%.1f%%',colors=colors,explode=explode, startangle=90, counterclock=False)
plt.show()
```

```python
wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':2}
v = [30, 25, 20, 13, 10, 2]
labels = ['python','java','c','c++','go','etc']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']

plt.pie(v,labels=labels, autopct='%.1f%%',colors=colors,startangle=90,wedgeprops=wedgeprops, counterclock=False)
plt.show()
```

```python
def custom_autopct(pct):
    return ('%.1f%%' % pct) if pct>=10 else ''

wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':2}
v = [30, 25, 20, 13, 10, 2]
labels = ['python','java','c','c++','go','etc']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']

plt.pie(v,labels=labels, autopct=custom_autopct,colors=colors,startangle=90,wedgeprops=wedgeprops, pctdistance=0.7)
plt.show()
```
