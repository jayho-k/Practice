# 04.style

```python
# 이것 또한 matplotlib에서 공식문서 확인
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
# 그래프의 색깔 등등을 바꾸는 것
plt.plot(x,y)
plt.show()
```

```python
# 선을 두껍게 만들기
plt.plot(x,y, linewidth=5)
```

## marker

 줄려서 표현 가능 하다 \
 plt.plot(x,y, marker='X',mfc='red', ms=10, mec='blue', ls=':')

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y, marker='o', linestyle="None")
```

```python
plt.plot(x,y, marker='v', markersize=10)
```

```python
plt.plot(x,y, marker='X', markersize=10)
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor='red',markerfacecolor = 'yellow')
```

## line style

```python
plt.plot(x,y,linestyle=':')
```

```python
plt.plot(x,y,linestyle='--')
```

```python
plt.plot(x,y,linestyle='-.')
```

## color

```python
plt.plot(x,y, color='red')
plt.plot(x,y, color='r')
plt.plot(x,y, color='g')
plt.plot(x,y, color='b')
# 모두 가능하다
```

## format 한번에 표현가능

```python
plt.plot(x,y, 'ro--')
```

```python
# line을 지정해주지 않으면 line이 없어진다.
plt.plot(x,y, 'ro')
```

## Transparency (alpha= )

```python
plt.plot(x,y, 'ro--',alpha = 0.2)
```

## graph size

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10)) 
plt.plot(x,y)
```
