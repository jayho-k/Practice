# Matplotlib

다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있다

```python
import matplotlib.pyplot as plt
```

## 기본 그래프

```python
x = [1,2,3]
y = [2,4,8]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show() # 이런식으로 하면 그래프만 보여주게 된다.
```

## Title 설정

```python
# 제목을 설정해주는 방법이다
plt.plot(x,y)
plt.title('Line Graph')
```

```python
# 제목을 한글, 마이너스폰트 깨지는 것을 막기 위해서
# 이 3줄은 항상 적기
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕
matplotlib.rcParams['font.size'] = 15
# 마이너스 표현하기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용가능한 폰트를 확인할 수 있다.
[f.name for f in fm.fontManager.ttflist]
# 폰트 이름만 출력하는 것
```

```python
plt.plot(x,y)
plt.title('그래프')
```

```python
# 마이너스를 표현하기
plt.plot([-1,0,1],[-5,0,3])
plt.show()
```
